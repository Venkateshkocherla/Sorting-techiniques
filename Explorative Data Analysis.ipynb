# EXPLORATORY DATA ANALYSIS

### Explorative Data Analysis is a process where one learns about the data, forms insights and identifies important columns (features) that can be user to tell a story or later formulate a ML problem.

```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

#### LOADING FLIGHTS DATA SET

```python
f = sns.load_dataset("flights")
f.head()
f.tail()
```

# LINEPLOT

Plotting **lineplot** between year and month and taking color parameter as **passengers**

```python
sns.lineplot(y="year",x="month",data=f,hue="passengers")
plt.show()
```

From the above lineplot we can conclude that the number of passengers travelling in flight increased from 1950's to 1960's

#### Loading **planets** data set 

```python
k = sns.load_dataset("planets")
k.head(10)
```

```python
k.describe()
```

# BOXPLOT

plotting a **boxplot** between year and distance from the above data set

```python
plot =sns.boxplot(x="distance",y="year",data=k)
plt.show()
```
